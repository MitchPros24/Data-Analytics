```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor



df = pd.read_csv("housing_train.csv")


numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

categorical_cols = df.select_dtypes(include=["object"]).columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")


```

```python
plt.figure()
plt.scatter(df["GrLivArea"], df["SalePrice"])
plt.xlabel("Living Area")
plt.ylabel("Sale Price")
plt.title("Living Area vs Sale Price")
plt.show()
```

<h2>Insiht:</h2>

<h3>1.Houses below 900 sq ft rarely exceed high prices, showing that very small homes have a strict value ceiling regardless of location or condition.
<br/>
2.The strongest price acceleration begins after 1,600 sq ft, showing that once houses pass a comfort size threshold, buyers pay significantly more.

3.Several large houses still appear at low prices, proving that size alone cannot override poor quality or weak market location.</h3>

```python
quality_price = df.groupby("OverallQual")["SalePrice"].mean()

plt.figure()
quality_price.plot(kind="line", marker="o")
plt.xlabel("Overall Quality")
plt.ylabel("Average Sale Price")
plt.title("Overall Quality vs Average Sale Price")
plt.show()
```

<h2>Insight:</h2>

<h3>4.The largest average price jump happens between quality level 6 and 8, meaning mid-to-high upgrades give the biggest return on investment.
<br/>
5.Quality levels 13 remain tightly compressed in price, showing that very low-quality homes are capped regardless of size.

6.Houses rated 9 and above show unstable pricing, which suggests luxury pricing is strongly influenced by features not included in the dataset.</h3>

```python
plt.figure()
plt.scatter(df["YearBuilt"], df["SalePrice"])
plt.xlabel("Year Built")
plt.ylabel("Sale Price")
plt.title("House Age vs Sale Price")
plt.show()
```

<h2>Insight:</h2>

<h3>7.Houses built after 2000 consistently maintain higher price floors, meaning depreciation slows significantly once modern standards apply.

8.Older homes built before 1980 show extreme price spread, proving that renovation quality causes large value differences in old properties.</h3>

```python
plt.figure()
plt.scatter(df["GarageArea"], df["SalePrice"])
plt.xlabel("Garage Area")
plt.ylabel("Sale Price")
plt.title("Garage Area vs Sale Price")
plt.show()
```

<h2>Insight:</h2>

<h3>9.Garage sizes below 350 sq ft show weak price impact, meaning small garages add convenience but do not meaningfully raise resale value.

10.Price only increases sharply when garage area and house size grow together, proving that oversized garages alone do not improve property value.</h3>
